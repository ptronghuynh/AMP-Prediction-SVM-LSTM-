# SVM_combined_internal

Internal SVM-combined analysis notebook. The model uses fixed-length sequence-index features together with the 126 GPSD descriptor columns and an RBF-kernel SVM.

```python
from pathlib import Path
import csv
import sys

ROOT = Path.cwd().resolve()
if ROOT.name == 'notebooks':
    ROOT = ROOT.parent
sys.path.insert(0, str(ROOT / 'scripts'))
features_path = ROOT / 'data' / 'Data_6kGoc' / 'features.csv'
with features_path.open(newline='', encoding='utf-8-sig') as handle:
    reader = csv.DictReader(handle)
    descriptor_cols = [c for c in reader.fieldnames if c not in {'ID', 'Label', 'Sequence'}]
    first_row = next(reader)
print('Feature file:', features_path)
print('Descriptor columns:', len(descriptor_cols))
print('First record label:', first_row['Label'])
```

```python
try:
    from sklearn.pipeline import Pipeline
    from sklearn.preprocessing import StandardScaler
    from sklearn.svm import SVC

    svm_combined = Pipeline([
        ('scale', StandardScaler()),
        ('svm', SVC(kernel='rbf', C=1.0, gamma='scale', probability=True, random_state=42)),
    ])
    print(svm_combined)
except ImportError as exc:
    print('Install scikit-learn from requirements.txt to instantiate the SVM pipeline:', exc)
```
